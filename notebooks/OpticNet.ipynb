```python
import h5py
import tensorflow as tf
from tensorflow import keras
```

```python
filepath = 'Optic_net-4_classes-Kermany2018.hdf5'
model = keras.models.load_model(filepath)
```

```python
weights=model.get_weights()
```

```python
names = [weight.name for layer in model.layers for weight in layer.weights]
for name, w in zip(names, weights):
    print(name, w.shape)
```

```python
pytorch_model = OpticNet(3,4)
for name, param in pytorch_model.named_parameters():
    print(name, param.shape)
```

```python
len(list(pytorch_model.named_parameters()))
```

```python
i = 0
list_keys = []
state_dict = pytorch_model.state_dict() 
for k in state_dict:
    if 'num_batches_tracked' in k:
        continue
    i += 1
    print(k, state_dict[k].shape)
    list_keys.append(k)


```

```python
new_state_dict = {}
import torch
for name, w, k in zip(names, weights, list_keys):
    if w.ndim==4:
        w = w.transpose((3,2,0,1))
    if w.ndim==2:
        w = w.transpose((1,0))
    print(name, k, w.shape, state_dict[k].shape)
    new_state_dict[k] = torch.from_numpy(w)
```

```python
pytorch_model.load_state_dict(new_state_dict)
```
