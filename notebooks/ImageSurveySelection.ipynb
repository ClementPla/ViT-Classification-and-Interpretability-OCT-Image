```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import tqdm
```

```python
attention_vit = 'Explained-ViT-AttentionRollout/'
lrp_vit = 'Explained-ViT-LRP/'
list_files = [f for f in os.listdir(lrp_vit) if not f=='.directory']
```

```python
data = {'image':[], 'rmse':[]}
df = pd.DataFrame(data)
for f in tqdm.tqdm(list_files):
    att = cv2.imread(os.path.join(attention_vit, f), cv2.IMREAD_GRAYSCALE)
    lrp = cv2.imread(os.path.join(lrp_vit, f),  cv2.IMREAD_GRAYSCALE)
    try:
        rmse = np.sqrt(((att-lrp)**2).mean())
    except TypeError:
        print(f)
        raise
    data = {'image':f, 'rmse':rmse}
    df = df.append(pd.Series(data), ignore_index=True, sort=True)
    
```

```python
    df['Label'] = df.apply(lambda row: row.image.split('-')[0], axis=1) 
  
```

```python
def select_imgs(df, label, nb_row=100):
    label_df = df.loc[lambda df: df['Label'] == label]
    label_df.sort_values(by='rmse', ascending=False, inplace=True)
    return label_df[:nb_row]
    
    
```

```python
dataframe =  select_imgs(df, 'DME')
dataframe = dataframe.append(select_imgs(df, 'DRUSEN')).append(select_imgs(df, 'CNV'))
```

```python
from shutil import copyfile
```

```python
out_folder = 'samples_survey'
if not os.path.exists(out_folder):
    os.makedirs(out_folder)
    
lrp_vit_folders = (lrp_vit, os.path.join(out_folder, 'Heatmap', 'LRP_ViT'))
att_vit_folders = (attention_vit, os.path.join(out_folder, 'Heatmap', 'Attention_ViT'))
dtd_resnet_folders = ('Explained-ResNet-DTD', os.path.join(out_folder, 'Heatmap', 'DTD_ResNet152'))
original_folder = ('/home/clement/Documents/database/CellData/OCT/test/',os.path.join(out_folder,'Images'))


def copy_contents(list_files, original_folder, destination_folder, test_folder=False):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for f in list_files:
        if test_folder:
            img_folder = os.path.join(original_folder, f.split('-')[0])
            copyfile(os.path.join(img_folder, f), os.path.join(destination_folder, f))
        else:
            copyfile(os.path.join(original_folder, f), os.path.join(destination_folder, f))
            
```

```python
copy_contents(dataframe['image'], *lrp_vit_folders)
copy_contents(dataframe['image'], *att_vit_folders)
copy_contents(dataframe['image'], *dtd_resnet_folders)
copy_contents(dataframe['image'], *original_folder, True)
```
